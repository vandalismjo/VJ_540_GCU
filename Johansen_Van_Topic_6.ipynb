### Problem 2 Part 1

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
%matplotlib inline
import skfuzzy as fuzz
from skfuzzy import control as ctrl

x1 = np.linspace(-10,10,50) 
x2 = np.linspace(-10,10,50) 

y = (np.sin(x1) * np.sin(x2))/(x1 * x2)
plt.plot(x1,y)
plt.show()

ax = plt.axes(projection='3d')
ax.scatter3D(x1, x2, y)
ax.plot3D(x1, x2, y)
plt.show()
```

```python
# New Antecedent/Consequent objects hold universe variables and membership functions
x1 = ctrl.Antecedent(np.arange(-10, 10, 1), 'x1')
x2 = ctrl.Antecedent(np.arange(-10, 10, 1), 'x2')
y = ctrl.Consequent(y, 'y')

# Auto-membership function population is possible with .automf(3, 5, or 7)
x1.automf(3)
x2.automf(3)

# Custom membership functions can be built interactively with a familiar Pythonic API
y['low'] = fuzz.trimf(y.universe, [0, 0, 0.5])
y['high'] = fuzz.trimf(y.universe, [0, 0.5, 1])
```

```python
# Set up training and testing data - sampled from uniform distribution
train_x1 = np.random.uniform(1, 6, 50) 
train_x2 = np.random.uniform(1, 6, 50) 
train_x3 = np.random.uniform(1, 6, 50) 

test_x1 = np.random.uniform(1.5, 5.5, 50) 
test_x2 = np.random.uniform(1.5, 5.5, 50) 
test_x3 = np.random.uniform(1.5, 5.5, 50) 

y_train = (1 + (train_x1)**0.5 + (train_x2)**(-1) + (train_x3)**(-1.5))**2
y_test = (1 + (test_x1)**0.5 + (test_x2)**(-1) + (test_x3)**(-1.5))**2

# New Antecedent/Consequent objects hold universe variables and membership functions
train_x1 = ctrl.Antecedent(np.arange(-10, 10, 1), 'train_x1')
train_x2 = ctrl.Antecedent(np.arange(-10, 10, 1), 'train_x2')
train_x3 = ctrl.Antecedent(np.arange(-10, 10, 1), 'train_x3')

test_x1 = ctrl.Antecedent(np.arange(-10, 10, 1), 'test_x1')
test_x2 = ctrl.Antecedent(np.arange(-10, 10, 1), 'test_x2')
test_x3 = ctrl.Antecedent(np.arange(-10, 10, 1), 'test_x3')

y_train = ctrl.Consequent(y_train, 'y_train')
y_test = ctrl.Consequent(y_test, 'y_test')

# Auto-membership function population is possible with .automf(3, 5, or 7)
train_x1.automf(3)
train_x2.automf(3)
train_x3.automf(3)
test_x1.automf(3)
test_x2.automf(3)
test_x3.automf(3)

# Custom membership functions can be built interactively with a familiar Pythonic API
y_train['low'] = fuzz.trimf(y_train.universe, [0, 0, 0.5])
y_train['high'] = fuzz.trimf(y_train.universe, [0, 0.5, 1])
y_test['low'] = fuzz.trimf(y_test.universe, [0, 0, 0.5])
y_test['high'] = fuzz.trimf(y_test.universe, [0, 0.5, 1])


# An adaptive neuro-fuzzy inference system or adaptive network-based fuzzy inference system (ANFIS) is 
# a kind of artificial neural network that is based on Takagi–Sugeno fuzzy inference system. The 
# technique was developed in the early 1990s. Since it integrates both neural networks and 
# fuzzy logic principles, it has potential to capture the benefits of both in a single framework. 
# Its inference system corresponds to a set of fuzzy IF–THEN rules that have learning capability 
# to approximate nonlinear functions. Hence, ANFIS is considered to be a universal estimator (Rajurkar, 2017).

## From the given data, rules, antecedents, and membership functions we can set up the ANFIS neural network
```
